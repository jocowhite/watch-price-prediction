## Download images from the shops

````python
def get_picture(row, folder):
    url = row[1]["Image URL"]
    code = row[1]["Code"]
    
    try:
        # Bild herunterladen und speichern
        image_filename = "data/watches/" + folder + f"/images/{code}.jpg"
        urllib.request.urlretrieve(url, image_filename)
        return "Bild gespeichert:" + image_filename
    except Exception as e:
        return "Fehler beim Herunterladen des Bildes:" +  url +  "Fehlermeldung:" +  str(e)
    
def download_images(df, folder):
    with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:

        future_row = [executor.submit(get_picture, row, folder) for row in df.iterrows()]

        for future in concurrent.futures.as_completed(future_row):
            try:
                print(future.result())
            except Exception as exc:
                print(f"Fehler: {exc}")
````

# Data Preperation

````python
## Helperfunctions

### Change the used filetype here
def add_jpg_suffix(string):
    return string + '.jpg'

### You can change the conversion rate here
def convert_usd_to_eur(price):
    return '%.2f' % (price * 1.0675)
````

````python
def format_data(df, drop_colums = None,
                      replace_stringparts = None, 
                      convert_price = None,
                      convert_to_usd = False,
                      select_category= None,
                      create_path_from_name = False,
                      reindex = None):

    #Choose a category from the Shopify Shop Categories if the Shop has not only watches
    if select_category != None:
        df = df[df['Category'] == select_category]
    
    #Drop the columns we don't need for the Model
    if drop_colums != None:
        df = df.drop(drop_colums, axis=1)
    
    #Replace unnecessary string parts in the Price String of the Kaggle Data
    if replace_stringparts != None:
        for replace_object in replace_stringparts:
            df['price'] = df['price'].replace(replace_object , '', regex=True)
    
    #Convert the Price to a specific Datatype
    if convert_price != None:
        df["price"].astype(float)
        
    if convert_to_usd:
        df['price'] = df['price'].apply(convert_usd_to_eur)
        
    if create_path_from_name:
        df['image_name'] = df['Code'].apply(add_jpg_suffix)
        
    if reindex != None:
        df = df.reindex(columns=reindex)

    return df
````

````python
def remove_nonexisting(path_to_cleandata, path_to_images):
    df = pd.read_csv(path_to_cleandata, index_col=0)
    #Get all Filenames in Image Folder
    file_names = []
    for filename in os.listdir(path_to_images):
        if os.path.isfile(os.path.join(path_to_images, filename)): 
            file_names.append(filename)
            #Remove non existing Pictures from Dataframe
    for index, row in df.iterrows():
        if row['image_name'] not in file_names:
            df.drop(index, inplace=True) 
    
    df = df.to_csv(path_to_cleandata)
````

## Clean DS1 Kaggle Data

````python
df_kaggle = pd.read_csv("data/metadata/metadata.csv") 
df_kaggle = format_data(df_kaggle,
                        replace_stringparts=["\$", " ", "'", '"', ","],
                        convert_price=True,
                        drop_colums=["brand", "name"],
                        reindex=['image_name', 'price'],
                       )
df_kaggle.to_csv("data/watches/ds1_kaggle/cleandata.csv")
remove_nonexisting(path_to_cleandata="data/watches/ds1_kaggle/cleandata.csv",
                   path_to_images="data/watches/ds1_kaggle/images"
                  )
df_kaggle.head()
````

````python
df_marc = pd.read_csv("data/metadata/marc_products.csv")
df_marc = format_data(df_marc,
                        select_category='Uhren',
                        create_path_from_name=True,
                        convert_to_usd=True,
                        reindex=['image_name', 'price'],
                        drop_colums=['Name', 'URL', 'Image URL','Collection', 'Category', 'Variant Name', 'In Stock'])
df_marc.to_csv("data/watches/ds2_marc/cleandata.csv")
remove_nonexisting(path_to_cleandata="data/watches/ds1_kaggle/cleandata.csv",
                   path_to_images="data/watches/ds1_kaggle/images"
                  )
df_marc.head()
````

````python
df_crown = pd.read_csv("data/metadata/crown_products.csv")
df_crown = format_data(df_crown,
                        create_path_from_name=True,
                        reindex=['image_name', 'price'],
                        drop_colums=['Name', 'URL', 'Image URL','Collection', 'Category', 'Variant Name', 'In Stock'])
df_crown.to_csv("data/watches/ds3_crown/cleandata.csv")
remove_nonexisting(path_to_cleandata="data/watches/ds2_marc/cleandata.csv",
                   path_to_images="data/watches/ds2_marc/images"
                  )
df_crown.head()
````
